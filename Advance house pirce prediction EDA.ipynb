# House price prediction,Exploratory data analysis(EDA)

## Main aim is to understand more about the data

Before going to train any model it's very important to analyze the data. Initially we have raw data and in that dataset 
there are lots of NaN or missing value are present. In some of the feature there is large no of missing value are present. 
so that paricular feature may drop but before dropping any feature from the dataset 1st check, is there any relationship between the output feature with respect to that particular feature which has NaN value.
so all the information is gather from the Exploratory data analysis.

#  Missing value find out
from the seaborn library we can visualize the nan value of a particular column.


```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
# find out the feature, containing nan value in house price prediction dataset
```

```python
feature_with_na = [feature  for feature in df.columns  if  df[feature].isnull().sum()>1]
```

```python
for features in feature_with_na:
    print(features,np.round(df[features].isnull().mean(),4))
```

```python
## since there are lots of nan value are present in the dataset so we have
## to check is there any relationship between sales price and null value column
```

```python
for feature_name in feature_with_na:
    data = df.copy()
    data[feature_name] = np.where(data[feature_name].isnull(),1,0) ## ie if mising value is present then use 1 otherwise we choose 0
    sns.barplot(data = data,x = data[feature_name],y = data['SalePrice'])
    plt.title(feature_name)
    plt.show()
```

```python
# In any given dataset, the features contain different different type of data such as numerical,categorical,etc
# so find out the features which contain numerical data and categorical data.
```

```python
numerical_variable = [feature  for feature in df.columns   if df[feature].dtype !='O']
```

```python
len(numerical_variable)
```

```python
df[numerical_variable]
```

```python
##Date and time variable also called as temproal varibale. 
year_feature = [feature  for feature in df.columns   if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
df[year_feature]
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('yearsold')
plt.ylabel('salesprice')
plt.title('houseprice vs yearsold ')                     
```

```python
## Now checking is there price increasing with respect to year?

for feature in year_feature:
    if feature!='YrSold':
        data = df.copy()
        data[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('saleprice')
        plt.show()

    
```

```python
#as the yearBuilt increases, the selling price of house will decreases.
```

```python
discrete_feature = [feature  for feature in numerical_variable  if len(df[feature].unique())<25 and feature not in year_feature+['Id']]
```

```python
print('discreate variable count: {}'.format(len(discrete_feature)))
```

```python
##lets relationship between sales price and discrete feature
for feature in discrete_feature:
    data=  df.copy()
    sns.barplot(data = data,x = data[feature],y = data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('sales')
    plt.show()
    
```

```python
# now calculating the continuous variable 
continuous_variable = [feature  for feature in numerical_variable  if feature not in discrete_feature+year_feature+['Id']]
```

```python
for feature in continuous_variable:
    data = df.copy()
    sns.histplot(data= data[feature],bins = 25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```
